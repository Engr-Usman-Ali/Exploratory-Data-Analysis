```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Exploratory-Data-Analysis/Iris.csv')
print(data.head())
```

# **Univariate Analysis**

```python
# Set the style for seaborn
sns.set(style="whitegrid")
data.hist(bins=20, figsize=(12, 10))
plt.suptitle('Histograms of Iris Dataset Features')
plt.show()
```

```python
# Box plots for each feature
plt.figure(figsize=(12, 8))
for i, column in enumerate(data.columns[1:5], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='Species', y=column, data=data)
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```

# **Bivariate and Multivariate Analysis**

```python
# Pair plot to visualize relationships
sns.pairplot(data, hue='Species', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

# **Correlation Heatmap**

```python
plt.figure(figsize=(10, 8))
correlation_matrix = data.iloc[:, 1:5].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Iris Dataset Features')
plt.show()
```

# **Summary of Key Insights**
* **Species Differentiation:**
 Iris-setosa can be easily distinguished from Iris-versicolor and Iris-virginica based on petal dimensions.
Feature Relationships: Petal dimensions are more informative for species classification than sepal dimensions.
* **Correlation:**
 High correlation between petal length and width suggests they may be redundant features in predictive modeling.
* **Visualizations:**
 Effective in revealing patterns and relationships that may not be immediately obvious from raw data.

**Conclusion:**  The EDA of the Iris dataset reveals significant insights into the relationships between features and the distinct characteristics of different species, which can be leveraged for further analysis or predictive modeling.
